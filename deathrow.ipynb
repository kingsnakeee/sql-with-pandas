```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
deathrow = pd.read_csv('/Users/User/Desktop/DA Course (Cliff)/Datasets/tx_deathrow_full.csv',encoding = 'utf-8')
```

```python
deathrow.columns = ['execution','date_of_birth','date_of_offence','highest_education_level','last_name','first_name',
                        'tdcj_number','age_of_execution','date_received','execution_date','race','county','eye_color',
                   'weight','height', 'native_county', 'native_state', 'last_statement']
```

```python
db_connection = sqlite3.connect('/Users/User/Desktop/DA Course (Cliff)/SQL practice/deathrow.db.sqlite')

deathrow.to_sql(name = 'deathrow', con = db_connection, if_exists = 'replace', index = False)
```
